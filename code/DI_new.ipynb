```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.model_selection import train_test_split
from sklearn.svm import LinearSVC
from sklearn.linear_model import RidgeClassifier, Perceptron
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import classification_report

# Load the CSV file
df = pd.read_csv('all')
df = pd.read_csv('all',sep=';', header=None,names=["article","lang"])#,error_bad_lines=False)
filtered_df = df[df['lang'] != 'lang']
f_df = filtered_df[filtered_df['lang'] != 'lang']


texts = f_df['article'].values
labels = f_df['lang'].values
# Convert raw text into numerical representation
vectorizer = CountVectorizer()
X = vectorizer.fit_transform(texts)

# Split dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, labels, test_size=0.2)

# Train linear SVC model
svc_model = LinearSVC()
svc_model.fit(X_train, y_train)

# Evaluate performance
y_pred = svc_model.predict(X_test)
print('Linear SVC:')
print(classification_report(y_test, y_pred))

# Train Ridge Classifier model
ridge_model = RidgeClassifier()
ridge_model.fit(X_train, y_train)

# Evaluate performance
y_pred = ridge_model.predict(X_test)
print('Ridge Classifier:')
print(classification_report(y_test, y_pred))

# Train Perceptron model
perceptron_model = Perceptron()
perceptron_model.fit(X_train, y_train)

# Evaluate performance
y_pred = perceptron_model.predict(X_test)
print('Perceptron:')
print(classification_report(y_test, y_pred))

# Train Naive Bayes model
nb_model = MultinomialNB()
nb_model.fit(X_train, y_train)

# Evaluate performance
y_pred = nb_model.predict(X_test)
print('Naive Bayes:')
print(classification_report(y_test, y_pred))

```
